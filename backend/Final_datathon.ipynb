```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
# Load the dataset
df = pd.read_csv("smart_grid_energy_dataset_5000_rows.csv")

# View first few rows
df.head()
```

```python
df.info()
df.describe()
```

FOR SUPPLY

```python
# Features influencing renewable generation
X_supply = df[[
    "solar_irradiance",
    "cloud_cover",
    "wind_speed",
    "temperature",
    "humidity"
]]

# Target variable (what we want to predict)
y_supply = df["total_supply_kWh"]
```

```python
# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_supply,
    y_supply,
    test_size=0.2,
    random_state=42
)
```

```python
# Initialize model
supply_model = RandomForestRegressor()

# Train model
supply_model.fit(X_train, y_train)
```

```python
# Predict supply on test data
predicted_supply = supply_model.predict(X_test)

# Calculate prediction error
print("Supply Prediction MAE:",
      mean_absolute_error(y_test, predicted_supply))
```

FOR DEMAND

```python
# Features affecting electricity demand
X_demand = df[[
    "hour",
    "day_of_week",
    "temperature",
    "num_ev_charging"
]]

# Target demand
y_demand = df["total_demand_kWh"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_demand,
    y_demand,
    test_size=0.2,
    random_state=42
)
```

```python
# Initialize model
demand_model = RandomForestRegressor()

# Train model
demand_model.fit(X_train, y_train)
```

```python
# Predict demand
predicted_demand = demand_model.predict(X_test)

# Measure error
print("Demand Prediction MAE:",
      mean_absolute_error(y_test, predicted_demand))
```

Predict Supply & Demand for Entire Dataset

```python
# Predict supply using trained model
df["predicted_supply"] = supply_model.predict(X_supply)

# Predict demand using trained model
df["predicted_demand"] = demand_model.predict(X_demand)
```

```python
# Calculate energy balance
df["energy_balance"] = df["predicted_supply"] - df["predicted_demand"]
```

```python
# Function deciding grid action
def grid_decision(balance):

    if balance > 0:
        return "Store Excess Energy"

    elif balance < 0:
        return "Use Grid Storage"

    else:
        return "Balanced"


# Apply rule to dataset
df["grid_action"] = df["energy_balance"].apply(grid_decision)
```

```python
df[[
    "predicted_supply",
    "predicted_demand",
    "energy_balance",
    "grid_action"
]].head(10)
```

```python
def ev_charging_scheduler(balance):

    if balance > 100:
        return "Allow EV Charging (Surplus renewable energy)"

    elif balance > -100:
        return "Limit EV Charging (Grid near balance)"

    else:
        return "Delay EV Charging (Prevent overload)"
```

```python
def self_healing_rerouting(balance):

    if balance > 200:
        return "Charge Battery Storage (Store excess renewable energy)"

    elif balance >= 0:
        return "Grid Stable (No rerouting required)"

    elif balance > -300:
        return "Use Battery Storage (Compensate deficit)"

    else:
        return "Activate Main Grid Backup (Critical deficit)"
```

```python
ev_action = ev_charging_scheduler(energy_balance)

rerouting_action = self_healing_rerouting(energy_balance)
```
